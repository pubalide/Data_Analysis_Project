```python
#import python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the data from the csv file
df = pd.read_csv('amazon.csv')
```

```python
#print first ten datas
df.head(10)
```

```python
#print last five datas
df.tail()
```

```python
df.info()
```

```python

df.shape
```

```python
df['rating'].describe()
```

```python
# check for missing values
df.isnull().sum()
```

```python
# null place
df[df.rating_count.isnull()]
```

```python
#  removing rows with null values
df.dropna(subset=['rating_count'],inplace=True)
df.isnull().sum()
        
```

```python
#check duplicate rows
print("Number of duplicate rows:")
df.duplicated().sum()
```

```python
# Replace string value and change data type
df['discounted_price'] = df['discounted_price'].str.replace('‚Çπ', '')
df['discounted_price'] = df['discounted_price'].str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].str.replace('‚Çπ', '')
df['actual_price'] = df['actual_price'].str.replace(',', '').astype(float)
df.head()
```

```python
#convert discount percentage to float
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(float)
df['discount_percentage'] = df['discount_percentage']/100
df.head()
```

```python
#counting rating
df['rating'].value_counts()
```

```python
#print place
df[df['rating'] == '|']
```

```python
#drop the row
row_to_edit = 1279
column_to_edit = 'rating'
df.at[row_to_edit, column_to_edit] = None 
df['rating'].value_counts()
```

```python
#convert rating count to float and print five rows
df['rating_count'] = df['rating_count'].str.replace(',', '').astype(float)
df.head()
```

```python
#rating count column and convert to float 
df['rating']=df['rating'].astype('float')
df.info()
```

```python
#count category
df['category'].value_counts()
```

```python
#split category column to two columns
df['main_category'] = df['category'].astype(str).str.split('|').str[0]
df['sub_category'] = df['category'].astype(str).str.split('|').str[-1]
df.head()
```

```python
#value count
df['main_category'].value_counts()
```

```python
#value count
df['sub_category'].value_counts()
```

```python
# Create new data frame with selected columns for data analysis
df1 = df[['product_id', 'product_name', 'category', 'discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count','main_category',
         'sub_category']].copy()
df1.head()
```

```python
# calculate total selling price price
total_selling_price=df1['discounted_price'].sum()
print(total_selling_price)
```

```python
#statistical data for numeric columns
df1.describe().T
```

```python
#statistical data for object columns
df1.describe(include=object).T
```

```python
#create pivot table
pd.pivot_table(df1,index=['main_category','sub_category'],values=['product_id'],aggfunc=['count'])
```

```python
#Top main categories in reviews
df.groupby(['main_category'])['rating_count'].sum().nlargest(10)
```

```python
#create pivot table
pd.pivot_table(df1,index=['main_category','product_id'],values=['discount_percentage','rating'],aggfunc=[np.sum])
```

**Here from pivot table we get clear idea that which main category product has good rating and has good discount percentage.**

```python
#Correlation between numeric variables
corr = df1.select_dtypes('number').corr()
sns.heatmap(data = corr, annot = True)
plt.title('Correlation between numeric variables')
plt.show()
```

**Here we see that discount_price have good correlation to actual_price**

```python
df1['main_category'].value_counts().plot(kind="barh")
```

**From the above bar plot,it is shown that Electronics products are selling more than other product.**

```python
# Histogram
plt.figure(figsize=(8, 5))
plt.hist(df1['rating'],bins=20,color='skyblue',edgecolor='black')
plt.xlabel('rating')
plt.ylabel('count')
plt.title('rating by count')
plt.tight_layout()
plt.show()
```

**From this above chart, The rating range of most products is around 4.0 and 4.5 and no products rated lower than 2.0.** 

```python
plt.figure(figsize=(8, 5))
plt.hist(df1['discount_percentage'],bins=20,color='pink',edgecolor='black')
plt.xlabel('discount_percentage')
plt.ylabel('category')
plt.title('category by discount percentage')
plt.tight_layout()
plt.show()
```

**From this above chart, most of the category have 40% to 60% discount.**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df ,y='actual_price' , x='discount_percentage')
plt.title('Prices vs Discounts')
plt.show()
```

**from the above scatter plot, most discounts are getting where the actual price is less than 20000. one product is present here which price is around
140000 which have around 40% discount**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df ,y='discounted_price' , x='rating')
plt.title('Prices vs Rating')
plt.show()
```

**from the above scatter plot, most prices were in range less than 60000 where the people are giving rating most.**

```python
df1.groupby(['sub_category'])['discount_percentage'].mean().nlargest(10).plot(kind='barh')
plt.title('Average discount percentage by sub categories')
plt.xlabel('Discount percentage')
plt.show()
```

**From the above bar chart, cableconnectionprotectors, Earpads and phonecharms are getting more discount.**

```python
#creating a line plot

rating_count = df1.groupby('rating')['rating_count'].sum().reset_index()

plt.plot(rating_count['rating'], rating_count['rating_count'], marker='o', label='rating')

plt.xlabel('rating')
plt.ylabel('rating_count')
plt.grid(True)
plt.title('rating count')
plt.legend()

plt.show()
```

**According to the above graph,the highest rating is around 4.0 to 4.5.No rating is less than 2.0.Most of the product has rating around 2.5 to 4.0.**

```python
df1.groupby(['discount_percentage'])['rating'].sum().nlargest(10).plot(kind='barh')
plt.title('rating by discounted percentage')
plt.xlabel('rating')
plt.show()
```

**From the above bar plot, it is shown that highest rating product has 50% discount**

```python
df1.groupby(['sub_category'])['rating'].sum().nlargest(5).plot(kind='pie', autopct='%1.1f%%')
plt.title('rating distribution')
plt.show()
```

**From the above pie chart, it is shown that highest rating product is USBCables among the sub category.**

```python
df1.groupby(['rating'])['rating_count'].sum().nlargest(5).plot(kind='pie', autopct='%1.1f%%')
plt.title('rating count')
plt.show()
```

**From the above pie chart, it is shown that highest rating is 4.1 which is given by most of the customers.**

```python
df1.groupby('main_category')['rating'].sum().sort_values(ascending=False).head(10).plot(kind='bar',title='main category vs rating',y='category')
```

**From the above bar plot, it is shown that highest rating product is from electronics among the main category.**

```python
sns.countplot(x='rating', data=df1)
```

**From the above chart, it is shown that highest rating is 4.1 which is given by most of the customers.**

```python
df1.groupby(['main_category'])['discount_percentage'].sum().nlargest(4).plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of discount percentage')
plt.show()
```

**From the above pie chart, it is shown that Electronics products are getting highest disccount price .**
